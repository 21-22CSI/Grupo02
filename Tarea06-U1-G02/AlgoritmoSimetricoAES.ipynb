```python
import time
import unicodedata
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes

def lee_archivo(file):
  inicio = time.time()
  archivo = open(f"txt{file}.txt","r")
  data = archivo.read
  print(archivo)
  fin = time.time()
  #print(archivo.read())
  print(f"Tiempo de lectura de {file}.txt: {fin-inicio}")
  return archivo
  archivo.close()

arch1 = lee_archivo(250)
arch2 = lee_archivo(500)
arch3 = lee_archivo(1000)
arch4 = lee_archivo(2000)
arch5 = lee_archivo(5000)
arch6 = lee_archivo(10000)
```

2da etapa, Genera las claves de cifrado

```python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
        ("ñ", "n"),
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s
```

```python
def genera_key(archivo_):
  inicio = time.time()
  data = archivo_.read()
  data = normalize(data)
  #data_cast = bytes(data, 'UTF-8')
  data_cast = data.encode()
  key = get_random_bytes(32)
  cipher_aes = AES.new(key, AES.MODE_EAX)
  print(key)
  fin = time.time()
  print(f"Tiempo de generar la key: {fin-inicio}")
  return cipher_aes, data_cast, key
  
cifrado_aes1, data_conv1, k1 = genera_key(arch1)
cifrado_aes2, data_conv2, k2 = genera_key(arch2)
cifrado_aes3, data_conv3, k3 = genera_key(arch3)
cifrado_aes4, data_conv4, k4 = genera_key(arch4)
cifrado_aes5, data_conv5, k5 = genera_key(arch5)
cifrado_aes6, data_conv6, k6 = genera_key(arch6)
```

3ra etapa, cifrar el texto

```python
def cifra_aes(data_conv, cifrado_aes):
  inicio = time.time()
  ciphertext_aes, tag = cifrado_aes.encrypt_and_digest(data_conv)
  #print(ciphertext_aes)
  fin = time.time()
  print(f"Tiempo de cifrado: {fin-inicio}")
  return ciphertext_aes, tag

text_cirfrado1, t1 = cifra_aes(data_conv1, cifrado_aes1)
text_cirfrado2, t2 = cifra_aes(data_conv2, cifrado_aes2)
text_cirfrado3, t3 = cifra_aes(data_conv3, cifrado_aes3)
text_cirfrado4, t4 = cifra_aes(data_conv4, cifrado_aes4)
text_cirfrado5, t5 = cifra_aes(data_conv5, cifrado_aes5)
text_cirfrado6, t6 = cifra_aes(data_conv6, cifrado_aes6)
```

4ta etapa, descifrar el texto

```python
def decifra_aes(text_cirfrado_, key, cifrado_aes, tag):
  inicio = time.time()
  cipher_aes1_1 = AES.new(key, AES.MODE_EAX, nonce=cifrado_aes.nonce)
  decrypth_text_aes = cipher_aes1_1.decrypt(text_cirfrado_)
  fin = time.time()
  print(f"Tiempo de descifrado: {fin-inicio}")
  try:
    cipher_aes1_1.verify(tag)
    print("Autentico: ", decrypth_text_aes)
  except ValueError:
    print("Key incorrecta o archivo daniado")

decifra_aes(text_cirfrado1, k1, cifrado_aes1, t1)
decifra_aes(text_cirfrado2, k2, cifrado_aes2, t2)
decifra_aes(text_cirfrado3, k3, cifrado_aes3, t3)
decifra_aes(text_cirfrado4, k4, cifrado_aes4, t4)
decifra_aes(text_cirfrado5, k5, cifrado_aes5, t5)
decifra_aes(text_cirfrado6, k6, cifrado_aes6, t6)
```
