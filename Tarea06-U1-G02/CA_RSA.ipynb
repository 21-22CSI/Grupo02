```python
#buscamos una fuente segura de aleatoriedad, no usar nuestros propios algoritmos.
#usar generadores aleatorios criptograficamente seguros ej: Crypto
import Crypto
from Crypto.PublicKey import RSA
from timeit import default_timer as timer
#numero random 
random_generator = Crypto.Random.new().read

#private_key = private_key.exportKey(format='DER')
#public_key = public_key.exportKey(format='DER')
#private_key = binascii.hexlify(private_key).decode('utf8')
#public_key = binascii.hexlify(public_key).decode('utf8')
```

```python
def leer_Texto(palabras):
  array = []
  with open('txt'+str(palabras)+'.txt') as f:
      while True:
          data = f.read(64)
          if not data: return array
          array.append(data)
```

```python
from Crypto.Cipher import PKCS1_OAEP
#para cifrar usamos la llave publica
def encriptar(archivo, clave_publica):
  cipher = PKCS1_OAEP.new(clave_publica)
  array = []
  for pedazo in archivo:
      array.append(cipher.encrypt(pedazo.encode()))
  return array
```

```python
#para descifrar usamos la llave privada
def desencriptar(archivo, clave_privada):
  cipher = PKCS1_OAEP.new(clave_privada)
  decodificado = ""
  for pedazo in archivo:
      decodificado += cipher.decrypt(pedazo).decode()
  return decodificado
```

# 250 palabras

```python
start= timer()
mensaje = leer_Texto(250)
end= timer()
time= end-start
print('Tiempo en leer archivo:', time)
```

```python
#generamos llave privada
start= timer();
private_key = RSA.generate(1024,random_generator)
#con la llave privada generamos la llave publica
public_key = private_key.publickey()
end= timer()
time= end-start;
print("Tiempo en generar claves: ", time)
print(private_key)
print(public_key)
```

```python
start= timer()
mensaje_encriptado = encriptar(mensaje, public_key)
end= timer()
time= end-start
print("Tiempo en cifrar texto: ", time)
```

```python
for pedazo in mensaje_encriptado:
  print(pedazo)
```

```python
start= timer()
mensaje_descencriptado = desencriptar(mensaje_encriptado, private_key)
end= timer()
time= end-start
print("Tiempo en descifrar texto: ", time)
```

```python
print(mensaje_descencriptado)
```
