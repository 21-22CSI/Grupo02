```python
from elgamal.elgamal import Elgamal
from timeit import default_timer as timer
```

```python
def leer_Texto(palabras):
  array = []
  with open('txt'+str(palabras)+'.txt') as f:
      while True:
          data = f.read(256)
          if not data: return array
          array.append(data)
```

```python
def encriptar(archivo, clavePublica):
  array = []
  for pedazo in archivo:
      array.append(Elgamal.encrypt(pedazo.encode(), clavePublica))
  return array
```

```python
def desencriptar(archivo, clavePrivada):
  decodificado = ""
  for pedazo in archivo:
      decodificado += Elgamal.decrypt(pedazo, clavePrivada).decode()
  return decodificado
```

# 250 palabras

```python
inicio = timer()
archivo = leer_Texto(250)
fin = timer()
tiempoTotal = fin-inicio
print("tiempo de leer archivo: ")
print(tiempoTotal)
```

```python
inicio = timer()
clave_publica, clave_privada = Elgamal.newkeys(256)
fin = timer()
tiempoTotal = fin-inicio
print("tiempo de ejecutar las claves: ")
print(tiempoTotal)
```

```python
print(clave_publica)
print(clave_privada)
```

```python
inicio = timer()
mensaje_encriptado = encriptar(archivo, clave_publica)
fin = timer()
tiempoTotal = fin-inicio
print("tiempo de cifrar texto: ")
print(tiempoTotal)
```

```python
for pedazo in mensaje_encriptado:
        print(pedazo)
```

```python
inicio = timer()
mensaje_desencriptado = desencriptar(mensaje_encriptado, clave_privada)
fin = timer()
tiempoTotal = fin-inicio
print("tiempo de decifrar texto: ")
print(tiempoTotal)
```

```python
print(mensaje_desencriptado)
```
