```python
import time
from Crypto.Cipher import DES
from Crypto.Random import get_random_bytes

def lee_archivo(file):
  inicio = time.time()
  archivo = open(f"txt{file}.txt","r")
  fin = time.time()
  #print(archivo.read())
  print(f"Tiempo de apertura de {file}.txt: {fin-inicio}")
  return archivo
  archivo.close()
  

arch1 = lee_archivo(250)
arch2 = lee_archivo(500)
arch3 = lee_archivo(1000)
arch4 = lee_archivo(2000)
arch5 = lee_archivo(5000)
arch6 = lee_archivo(10000)
```

2da etapa, Genera las claves de cifrado

```python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
        ("ñ", "n"),
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s
```

```python
def genera_key(archivo_):
  inicio = time.time()
  data = archivo_.read()
  data = normalize(data)
  data_cast = bytes(data, 'UTF-8')
  key = get_random_bytes(8)
  iv = get_random_bytes(8)
  print(key)
  cipher = DES.new(key, DES.MODE_CFB, iv)
  fin = time.time()
  print(f"Tiempo de generar la key: {fin-inicio}")
  print(data)
  return cipher, data_cast, key, iv

cifrado_des1, data_conv1, k1, i_v1 =  genera_key(arch1)
cifrado_des2, data_conv2, k2, i_v2 =  genera_key(arch2)
cifrado_des3, data_conv3, k3, i_v3 =  genera_key(arch3)
cifrado_des4, data_conv4, k4, i_v4 =  genera_key(arch4)
cifrado_des5, data_conv5, k5, i_v5 =  genera_key(arch5)
cifrado_des6, data_conv6, k6, i_v6 =  genera_key(arch6)
```

3ra etapa, cifrar el texto

```python
def cifra_des(data_conv, cifrado_des):
  inicio = time.time()
  ciphertext_des = cifrado_des.encrypt(data_conv)
  fin = time.time()
  print(f"Tiempo de cifrado: {fin-inicio}")
  return ciphertext_des
  print(ciphertext_des)

text_cifrado1 = cifra_des(data_conv1, cifrado_des1)
text_cifrado2 = cifra_des(data_conv2, cifrado_des2)
text_cifrado3 = cifra_des(data_conv3, cifrado_des3)
text_cifrado4 = cifra_des(data_conv4, cifrado_des4)
text_cifrado5 = cifra_des(data_conv5, cifrado_des5)
text_cifrado6 = cifra_des(data_conv6, cifrado_des6)

```

4ta etapa, descifrar el texto

```python
def decifra_des(text_cifrado, key, iv):
  inicio = time.time()
  cipher_1 = DES.new(key, DES.MODE_CFB, iv)
  decrypth_text_des = cipher_1.decrypt(text_cifrado)
  fin = time.time()
  print(decrypth_text_des)
  print(f"Tiempo de descifrado: {fin-inicio}")

decifra_des(text_cifrado1, k1, i_v1)
decifra_des(text_cifrado2, k2, i_v2)
decifra_des(text_cifrado3, k3, i_v3)
decifra_des(text_cifrado4, k4, i_v4)
decifra_des(text_cifrado5, k5, i_v5)
decifra_des(text_cifrado6, k6, i_v6)
```
